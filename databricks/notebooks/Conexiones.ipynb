```python
import os
import subprocess

# Ver JAVA_HOME real
print("JAVA_HOME:", os.environ.get("JAVA_HOME"))

# Ver si java existe en esa ruta
result = subprocess.run(["which", "java"], capture_output=True, text=True)
print("which java:", result.stdout.strip())

# Ver versión
result2 = subprocess.run(["java", "-version"], capture_output=True, text=True)
print("java version:", result2.stderr.strip())

# Ver si la ruta hardcodeada existe
print("Path exists:", os.path.isdir("/usr/lib/jvm/java-17-openjdk-amd64"))
```

```python
import subprocess, os

# Ver si puede resolver maven central
result = subprocess.run(
    ["curl", "-s", "-o", "/dev/null", "-w", "%{http_code}", 
     "https://repo1.maven.org/maven2/"],
    capture_output=True, text=True
)
print("Maven Central HTTP:", result.stdout)

# Ver PYSPARK_SUBMIT_ARGS actual
print("PYSPARK_SUBMIT_ARGS:", os.environ.get("PYSPARK_SUBMIT_ARGS"))

# Ver versión de Spark instalada
result2 = subprocess.run(
    ["/usr/local/spark/bin/spark-submit", "--version"],
    capture_output=True, text=True
)
print("Spark version:", result2.stderr[:300])
```

```python
import os
# Temporalmente anula el PYSPARK_SUBMIT_ARGS
os.environ["PYSPARK_SUBMIT_ARGS"] = "pyspark-shell"

from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("LuisNotebookTest") \
    .getOrCreate()
print("Spark version:", spark.version)
spark.range(5).show()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master("local[*]") \
    .appName("LuisNotebookTest") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    .getOrCreate()

print("Spark version:", spark.version)
spark.range(5).show()
```
